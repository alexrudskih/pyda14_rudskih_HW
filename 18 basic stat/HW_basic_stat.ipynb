В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.

```python
#  проверочная ячейка

q1 = horse_8col['total protein'].quantile(0.25)
q3 = horse_8col['total protein'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
print(lower_bound)
upper_bound = q3 + (1.5 * iqr)
print(upper_bound)

# remove_outliers = horse_8col[horse_8col['pulse'].between(lower_bound, upper_bound, inclusive=True)]

outliers_low = horse_8col[horse_8col['total protein'] < lower_bound]
outliers_up = horse_8col[horse_8col['total protein'] > upper_bound]
outliers_concat = pd.concat([outliers_low, outliers_up])

outliers_concat
```

```python
def outliers(x, y, z):
    """ 
    Функция определяющая выбросы в DataFrame, где 1-ый аргумент - DataFrame, к которому применятеся функция, 
    2-ой аргумент - столбец, в котором осуществляется поиск выбросов,
    3-ий аргумент - изменяет результат работы функции: если поставить 0, то возвращаются ячейки с выбросами, 
    если 1, то возвращаются все ячейки, за исключением выбросов и ячеек, в которых в данном столбце стоит NaN. 
    
    """
    q1 = x[y].quantile(0.25)
    q3 = x[y].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr) 
    upper_bound = q3 + (1.5 * iqr)

# remove_outliers = horse_8col[horse_8col['pulse'].between(lower_bound, upper_bound, inclusive=True)]

    outliers_low = x[x[y] < lower_bound]
    outliers_up = x[x[y] > upper_bound]
    outliers_concat = pd.concat([outliers_low, outliers_up])
    
    remove_outliers = x[x[y].between(lower_bound, upper_bound, inclusive=True)]
    
    if z == 0:
        return outliers_concat
    else:
        return remove_outliers
```

## Задание 3. Работа с пропусками
Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе работы с пропусками по каждому столбцу, сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
# Посчитаем количество пропусков
for col in re_horse_8col.columns:
    pct_missing_0 = re_horse_8col[col].isnull().sum()
    pct_missing_1 = re_horse_8col[col].isnull().mean()
    print(f'{col} - {pct_missing_1 :.2%}, Количество пропусков: {pct_missing_0}')
```

```python
# В столбце surgery? пропущено одно значение, удалим данную строку
sur_re_horse_8col = re_horse_8col.dropna(subset = ['surgery?'])
```

```python
# заполним столбец 'rectal temperature' медианой, в зависимости от значения показателя 'surgery?' (наличия операции)
_sur_re_horse_8col = sur_re_horse_8col.copy()
_sur_re_horse_8col['rectal temperature'].fillna(sur_re_horse_8col.groupby('surgery?')['rectal temperature'].transform('median'), inplace = True)
```

```python
# Столбец 'temperature of extremities'. Данный столбец должен коррелироваться с ректальной температурой. 
# Округлим значения ректальной температуры.
temp_ex_sur_re_horse_8col = _sur_re_horse_8col.copy()
temp_ex_sur_re_horse_8col['rectal temperature'] = temp_ex_sur_re_horse_8col['rectal temperature'].round()

temp_ex_sur_re_horse_8col['temperature of extremities'].fillna(temp_ex_sur_re_horse_8col.groupby('rectal temperature')['temperature of extremities'].transform('median'), inplace = True)
# Округлим значения в столбце до целого, т.к. медиана в некоторых ячейках оказалась не целым числом
temp_ex_sur_re_horse_8col['temperature of extremities'] = temp_ex_sur_re_horse_8col['temperature of extremities'].round()
```

```python
# столбец pulse рассчитаем на основе surgery? и rectal temperature, т.к. у здоровой лошади, он не должен быть высоким, 
# также как и температура
pulse_ = temp_ex_sur_re_horse_8col.copy()

pulse_['pulse'].fillna(temp_ex_sur_re_horse_8col.groupby(['surgery?', 'rectal temperature'])['pulse'].transform('median'),inplace = True)
```

```python
# добавил в общую таблицу вернозаполненные столбцы pulse и temperature of extremities
rty = _sur_re_horse_8col.join(pulse_, lsuffix = '_0', rsuffix = '_1')
rty = rty[['surgery?_0', 'Age_0', 'rectal temperature_0', 'pulse_1', 'temperature of extremities_1', 'abdominal distension_0', 'packed cell volume_0', 'total protein_0']]
rty.columns = ['surgery?','Age','rectal temperature','pulse','temperature of extremities','abdominal distension','packed cell volume','total protein']
```

```python
# показатель abdominal distension' связан с хирургическим вмешательством (surgery). 
# Также определим его в зависимости от показателя temperature of extremities.
ab_dist = rty.copy()
ab_dist['abdominal distension'].fillna(rty.groupby(['surgery?','temperature of extremities'])['abdominal distension'].transform('median'), inplace = True)
ab_dist['abdominal distension'] = ab_dist['abdominal distension'].round()
```

```python
# определим 'packed cell volume' от 'abdominal distension'
pcv = ab_dist.copy()
pcv['packed cell volume'].fillna(ab_dist.groupby('abdominal distension')['packed cell volume'].transform('median'), inplace = True)
```

```python
#  'total protein' связан с обезвоживанием. Показатель, также как 'packed cell volume'. РАссчитаем в зависимости от него
tp = pcv.copy()
tp['total protein'].fillna(pcv.groupby('packed cell volume')['total protein'].transform('median'), inplace = True)
# медианой получилось заполнить не все пропуски. Осталось пустых 8 значений или 2.68%. 
# Заполним оставшиеся пропуски медианой по столбцу "total protein"
tp_ = tp.copy()
tp_['total protein'].fillna(tp['total protein'].median(), inplace = True)
```

```python
for col in tp_.columns:
    pct_missing_0 = tp_[col].isnull().sum()
    pct_missing_1 = tp_[col].isnull().mean()
    print(f'{col} - {pct_missing_1 :.2%}, Количество пропусков: {pct_missing_0}')
```
