# Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
base = pd.read_csv('ml-latest/ratings.csv')
base.head()
```

```python
base.info()
```

```python
len(base['movieId'])
```

```python
# оставил строки, в которых оценка 5
base_edit_5 = base.query('rating == {}'.format(5.0))
base_edit_5.head()
```

```python
# сгруппировал по фильмам оценки и посчитал количество строк
base_edit_sort = base_edit_5.groupby('movieId').count().sort_values('userId', ascending = False)
base_edit_sort.head()
```

```python
movies = pd.read_csv('ml-latest/movies.csv')
movies.head()
```

```python
# присоединил таблицу с наименованиями фильмов и жанрами
joined_table = base_edit_sort.merge(movies, on = 'movieId', how = 'left')
joined_table.head()
```

```python
# вывел только нужные столбцы
joined_table_filter = joined_table[['movieId', 'title', 'genres']]
joined_table_filter.head()
```

# Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('power.csv')
data.head()
```

```python
# Lithuania - Литва
# Latvia - Латвия
# Estonia - Эстония
countries = ['Latvia', 'Estonia', 'Lithuania']
# years = [2005, 2006, 2007, 2008, 2009, 2010]
categories = [4, 12, 21]
```

```python
# отфильтровал по странам
data_country = data.query(f'country in {countries}')
```

```python
#  отфильтровал по годам
data_year = data_country.query('{} <= year <= {}'.format(2005,2010))
```

```python
#  отфильтровал по категориям
data_category = data_year.query(f'category in {categories}')
```

```python
# убрал отрицательные значения
data_not_negative = data_category.query(f'quantity > {0}')
```

```python
# Сгруппировал по странам и получил сумму значений
data_ = data_not_negative.groupby('country').sum().add_prefix('total_').reset_index()
```

```python
# Итоговый результат, убрал ненужные столбцы
data_result = data_[ ['country', 'total_quantity' ] ]
data_result.head()
```
